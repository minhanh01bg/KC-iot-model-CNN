```python
df = pd.DataFrame()
cnt = 1201
dir = './password_attack_rename'
for i in range(cnt,len(os.listdir(dir))+1):
    df_temp = pd.read_csv(dir+'/'+str(i)+'.csv')
    df = pd.concat([df,df_temp])
    print(df.shape)
    print(cnt)
    cnt += 1
endl = len(os.listdir(dir))
df.to_csv(f'./concat_par/1201_{endl}.csv',index=False)
```
