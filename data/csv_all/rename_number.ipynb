```python
import pandas as pd
import numpy as np
import os
```

```python
dir = './ddos/'
co = pd.DataFrame()
cnt = 0
for file in os.listdir(dir):
    cnt += 1
    print(cnt)
    if not file.endswith('.csv'):
        continue
    t = pd.read_csv(dir+file)
    t = t[t['Label'] != 'Label']
    t.to_csv(f'./ddos_rename/{cnt}.csv', index=False)
    
```
